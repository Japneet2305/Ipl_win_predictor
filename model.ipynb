```python
match_df[['Team1', 'Team2']] = match_df['Teams'].str.split(' vs ', expand=True)
```

```python
match_df.drop(columns=['Teams'], inplace=True)
```

```python
teams=[
    'Royal Challengers Bangalore',
    'Mumbai Indians',
    'Kolkata Knight Riders',
    'Rajasthan Royals',
    'Chennai Super Kings',
    'Sunrisers Hyderabad',
    'Delhi Capitals', 
    'Punjab Kings',
    'Lucknow Super Giants',
    'Gujarat Titans'
]
```

```python
match_df['Team1']=match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2']=match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1']=match_df['Team1'].str.replace('Royal Challengers Bengaluru','Royal Challengers Bangalore')
match_df['Team2']=match_df['Team2'].str.replace('Royal Challengers Bengaluru','Royal Challengers Bangalore')

match_df['Team1']=match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2']=match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['Team1']=match_df['Team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['Team2']=match_df['Team2'].str.replace('Kings XI Punjab','Punjab Kings')

match_df['Team1']=match_df['Team1'].str.replace('Gujarat Lions','Gujarat Titans')
match_df['Team2']=match_df['Team2'].str.replace('Gujarat Lions','Gujarat Titans')
```

```python
match_df=match_df[match_df['Team1'].isin(teams)]
match_df=match_df[match_df['Team2'].isin(teams)]
```

```python
match_df=match_df[['Match_ID','Venue','Match_Winner','total_runs']]
```

```python
delivery.head()
delivery.rename(columns={'Match id': 'Match_ID'}, inplace=True)
```

```python
delivery_df = match_df.merge(delivery,on='Match_ID')
delivery.head()
```

```python
delivery_df = delivery_df[delivery_df['Innings No']==2]
```

```python
delivery_df.head()
```

```python
delivery_df['runs_needed']=delivery_df['total_runs'] - delivery_df['score']
```

```python
delivery_df
```

```python
delivery_df['over']=delivery_df['Ball No'].astype(int)
```

```python
delivery_df['ball']=(delivery_df['Ball No']-delivery_df['over'])*10
```

```python
delivery_df.head()
```

```python
delivery_df['ball_left']=120- ((delivery_df['over']*6)+delivery_df['ball'])
```

```python
delivery_df.head()
```

```python
delivery_df['ball_left']=delivery_df['ball_left'].astype(int)
```

```python
delivery_df.head()
```

```python
delivery_df
```

```python
delivery_df.columns
```

```python
delivery_df[['curr_runs', 'wickets']] = delivery_df['score/wicket'].str.split('/', expand=True).astype(int)
```

```python
delivery_df
```

```python
delivery_df['wickets_left']=10-delivery_df['wickets']
```

```python
delivery_df
```

```python
delivery_df
```

```python
delivery_df['rrr']=(delivery_df['runs_needed'].astype(float)/delivery_df['ball_left'])*6
```

```python
delivery_df
```

```python
delivery_df.rename(columns={'ball_left': 'balls_left'}, inplace=True)
```

```python
delivery_df.loc[delivery_df['runs_needed'] < 0, 'runs_needed'] = 0
```

```python
delivery_df['crr']=delivery_df['score'].astype(float)/(delivery_df['over']*6+delivery_df['ball'])
```

```python
delivery_df['rrr']=(delivery_df['runs_needed'].astype(float)/delivery_df['balls_left'])*6
```

```python
delivery_df
```

```python
delivery_df.loc[delivery_df['Match_Winner'] == delivery_df['Batting team'], 'result'] = 1
```

```python
delivery_df.loc[delivery_df['Match_Winner'] != delivery_df['Batting team'], 'result'] = 0
```

```python
delivery_df['result'] = delivery_df['result'].astype(int)
```

```python
final_df = delivery_df[['Batting team','Bowling team','Venue','runs_needed','balls_left','wickets_left','crr','rrr','result']]
```

```python
final_df['rrr'].fillna(0, inplace=True)
final_df['rrr'] = final_df['rrr'].replace([np.inf, -np.inf], 36)
```

```python
final_df = final_df.sample(final_df.shape[0])
final_df.replace([np.inf, -np.inf], np.nan, inplace=True)
final_df.dropna(inplace=True)
final_df.loc[final_df['rrr'] > 36, 'rrr'] = 36
final_df.drop(final_df[final_df['crr'] > 36].index, inplace=True)
final_df.loc[final_df['Batting team']=='Deccan Chargers', 'Batting team'] = 'Sunrisers Hyderabad'
final_df.loc[final_df['Bowling team']=='Deccan Chargers', 'Bowling team'] = 'Sunrisers Hyderabad'
final_df.loc[final_df['Batting team']=='Kings XI Punjab', 'Batting team'] = 'Punjab Kings'
final_df.loc[final_df['Bowling team']=='Kings XI Punjab', 'Bowling team'] = 'Punjab Kings'
final_df.loc[final_df['Batting team']=='Gujarat Lions', 'Batting team'] = 'Gujarat Titans'
final_df.loc[final_df['Bowling team']=='Gujarat Lions', 'Bowling team'] = 'Gujarat Titans'
final_df.loc[final_df['Batting team']=='Delhi Daredevils', 'Batting team'] = 'Delhi Capitals'
final_df.loc[final_df['Bowling team']=='Delhi Daredevils', 'Bowling team'] = 'Delhi Capitals'
final_df.loc[final_df['Batting team']=='Royal Challengers Bengaluru', 'Batting team'] = 'Royal Challengers Bangalore'
final_df.loc[final_df['Bowling team']=='Royal Challengers Bengaluru', 'Bowling team'] = 'Royal Challengers Bangalore'
```

```python

X = final_df.iloc[:,:-1]
y = final_df.iloc[:,-1]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
tr = make_column_transformer(
    (OneHotEncoder(sparse_output=False,drop='first'),['Batting team','Bowling team','Venue']),
    remainder='passthrough'
)
tr.fit_transform(final_df)
```

```python
pipe = Pipeline(steps=[
    ('step1', tr),
    ('step2', LogisticRegression(solver='newton-cg'))
])
X_train.describe()
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
